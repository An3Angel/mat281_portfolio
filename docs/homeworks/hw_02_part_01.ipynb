# **Visualización**

```python
fig, ax  = plt.subplots(figsize=(16, 8))
fig.suptitle('Estados', size = 20)

top_states = districts["state"].value_counts().index

# Colores
colors = sns.color_palette("Spectral", n_colors=len(top_states))

# Gráfico de barras
districts["state"].value_counts().loc[top_states].plot(kind='bar',
                                                       figsize=(16, 8),
                                                       color=colors
                                                       )
plt.show()
```

Podemos observar que los datos sin información predominan en la columna de estados, por lo que no podemos obviarlos, para una mejor observación se puede crear alguna relación entrelos district_id.

```python
fig, ax  = plt.subplots(figsize=(16, 8))
fig.suptitle('Localidades', size = 20)

top_locale = districts["locale"].value_counts().index

# Paleta de colores
colors = sns.color_palette("Spectral", n_colors=len(top_locale))

# Gráfico
districts["locale"].value_counts().loc[top_locale].plot(kind='bar',
                                                        figsize=(16, 8),
                                                        color=colors
                                                        )
plt.show()
```
